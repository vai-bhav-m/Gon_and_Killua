#### Making a copy and preserving original data set in df:

```python
df1 = df.copy()
```

#### Dropping UserId column:

```python
df1.drop(["userId"], axis=1, inplace= True)
```

```python
df1.head()
```

#### Changing gender column to a binary (1,0) scheme:

```python
df1['gender'] = df1['gender'].map({1:1, 2:0}).astype('uint8')
```

#### One hot encoding "tier" column as it is a categorical variable:

```python
df1.head()
```

```python
df1.tier.value_counts()
```

```python
df2 = pd.get_dummies(df["tier"])
df2.columns = ["tier_1","tier_2","tier_3"]
df2.head()
```

```python
df1 = pd.concat([df1.drop(["tier"], axis=1),df2], axis=1)
```
