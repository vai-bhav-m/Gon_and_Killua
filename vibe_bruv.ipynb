```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("data/train_age_dataset.csv")
df.head()
```

#### Making a copy and preserving original data set in df:

```python
df1 = df.copy()
```

#### Dropping UserId column:

```python
df1.drop(["userId"], axis=1, inplace= True)
```

```python
df1.head()
```

### Dealing with categorical variables:

```python
df_cat = df1[["gender", "tier", "age_group"]]
```

```python
df_cat['gender'] = df_cat['gender'].map({1:1, 2:0}).astype('uint8')
```

```python
df_cat["tier"].value_counts()
```

```python
df2 = pd.get_dummies(df_cat["tier"])
df2.columns = ["tier_1","tier_2","tier_3"]
df2.head()
```

```python
df_cat = pd.concat([df_cat.drop(["tier"], axis=1),df2], axis=1)
```

```python
df_cat.head()
```

### Dealing with numeric variables:

```python
df_num = df1.drop(["tier","gender","age_group"], axis=1)
df_num.head()
```

```python
Std = df_num.std()
Mean = df_num.mean()
df_num = (df_num - Mean)/Std
```

```python
df_num.describe()
```

### Combining the Categorical and Numeric parts

```python
df_comb = pd.concat([df_cat, df_num], axis=1)
df_comb.head()
```

### Shuffling the data:

```python
df_shuffled = df_comb.iloc[np.random.permutation(range(len(df_comb)))]
```

```python
df_shuffled.reset_index(drop=True, inplace=True)
df_shuffled.head()
```

### Separating the output:

```python
df_y = df_shuffled[["age_group"]]
```

```python
df_shuffled.drop(["age_group"], axis=1, inplace= True)
```

```python
df_shuffled.head()
```

```python
df_y.head()
```

#### Current relevant variables in memory:

Dataframes:
- df --> untouched dataset
- df_comb --> normalized dataset 
- df_shuffled --> shuffled dataset
- df_y --> output

Transformation:
- Std --> standard deviation of dataset
- Mean --> mean of dataset

### Converting data into numpy arrays and splitting into training and validation sets:

```python
X = df_shuffled.to_numpy()
y = pd.get_dummies(df_y["age_group"]).to_numpy()
```

```python
X.shape, y.shape
```

```python
X[0], y[:4]
```

```python
# Keeping 40,000 examples in validation set and remaining in training set
X_train = X[:-40000]
X_val = X[-40000:]
y_train = y[:-40000]
y_val = y[-40000:]
```

```python
X_train.shape, y_train.shape
```

### Using keras to build a multi-class classifier

```python
# !pip install --upgrade pip
# !pip install tensorflow
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(64, activation=tf.nn.relu, input_dim=27),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(4, activation=tf.nn.softmax)
])
```

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20, validation_data=(X_val, y_val))
```

### Making Predictions on Validation set and evaluating it

```python
import sklearn
from sklearn import metrics
```

```python
y_val_pred = model.predict(X_val)
y_val_pred.shape
```

```python
for i in range(y_val.shape[0]):
    y_val_pred[i,:] = (y_val_pred[i,:] == max(y_val_pred[i,:]))
val_predict = y_val_pred @ np.array([1,2,3,4]).reshape(4,1)
y_val = y_val @ np.array([1,2,3,4]).reshape(4,1)
```

```python
print(metrics.classification_report(y_val, val_predict, digits=4))
```
